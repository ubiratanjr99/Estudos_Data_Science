**Dia 3: Primeira Inspeção:** Como "sentir" os seus dados? 
<br>
Use .head(), .tail(), .info(), .shape e .describe() para ter uma visão geral do seu DataFrame. 

```python
import pandas as pd

caminho_csv = '../arquivos/vendas_loja.csv'
caminho_excel = '../arquivos/filmes_series.xlsx'
```

### .shape
Vai retornar (qtd_linha, qtd_colunas). 
<br> 
Usado para saber o tamanho da sua base dados (dataframe).

```python
df = pd.read_csv(caminho_csv, sep=';', decimal=',')
df.shape
```

### .describe()
Vai retornar a análise incial da sua base de dados (média, quantidade de registros de cada coluna, valor mínimo e máximo de cada coluna, desvio padrão, etc).  


```python
df.describe()
```

### .info()
Vai retornar as informações de tipo de dado de cada coluna.  


```python
df.info()
```

### .head()
Vai retornar aos primeiros registros da base de dados. Pode-se especificar quantas linhas deseja como parâmetro da função. 


```python
# Sem parâmetro retorna os 5 primeiros registros
df.head()

# Com parâmtros retorna a quantidade especificada. No caso abaixo foi 10.
df.head(10)
```

### .tail()
É o contrário do método "head()", vai retornar os últimos registros da base de dados. 


```python
df.tail()

df.tail(10)
```
