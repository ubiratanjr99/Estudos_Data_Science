```python
import pandas as pd
```

```python
caminho = '../arquivos/vendas_loja.csv'
```

```python
df = pd.read_csv(caminho, sep=';', decimal=',')
```

```python
df.head()
```

Exercício 1: Filtrar por Valor Numérico
<br>
Objetivo: Selecionar e mostrar apenas as vendas com Valor superior a R$ 1.000,00.

```python
df[df['valor_total']>1000].sort_values(by='valor_total',ascending=False).head()
```

Exercício 2: Filtrar por Texto Exato
<br>
Objetivo: Selecionar e mostrar todas as vendas do produto Celular.

```python
df['produto'].unique()
```

```python
df[df['produto'] == 'Smartphone']
```

Exercício 3: Filtrar Múltiplas Condições (OU)
<br>
Objetivo: Selecionar as vendas em que a Quantidade é maior que 3 OU o Produto é Laptop.

```python
df[(df['quantidade'] > 3) | (df['produto'] == 'Laptop')]
```

Exercício 5: Filtrar usando .isin()
<br>
Objetivo: Selecionar todas as vendas dos produtos Mouse ou Teclado de uma só vez.
<br>
(Dica: o método .isin() é usado para verificar se o valor de uma coluna está contido em uma lista).

```python
df[df['produto'].isin(['Mouse', 'Teclado'])]
```
